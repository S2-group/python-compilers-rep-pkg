```python
import glob, os
import pandas as pd
import numpy as np
import seaborn as sns
from itertools import product
import matplotlib.pyplot as plt
import utils
import statistics
```

## Instructions
Set the global variable called SERVER to 'NUC' to obtain the data analysis for the NUC  or to 'server' for the data analysis of the server

```python
SERVER = 'server'
DATAPATH = f"../data/{SERVER}_data.csv"
```

```python
METRIC = 'TOTAL_ENERGY'
```

```python
data = pd.read_csv(DATAPATH)
utils.clean_and_format_df(data)
```

```python
# Convert to KJ and add Total Energy
data['DRAM_ENERGY (J)'] = data['DRAM_ENERGY (J)'] / 1000
data['PP0_ENERGY (J)'] = data['PP0_ENERGY (J)'] / 1000
data['TOTAL_ENERGY'] = data['PACKAGE_ENERGY (J)'] + data['DRAM_ENERGY (J)']
data['CACHE_ENERGY'] = data['PACKAGE_ENERGY (J)'] - data['PP0_ENERGY (J)']

data['PROCESS_MEMORY'] =  data['PROCESS_MEMORY'] / 1000000000
#data['MEMORY_USAGE'] = (data['PROCESS_MEMORY']) / (data["TOTAL_MEMORY"] / 1000000000) * 100

data.rename(
    columns={'PACKAGE_ENERGY (J)': 'PACKAGE_ENERGY (KJ)', 'DRAM_ENERGY (J)': 'DRAM_ENERGY (KJ)', 'PP0_ENERGY (J)': 'PP0_ENERGY (KJ)' }, inplace=True
)
```

```python
data = data.dropna(subset=[METRIC])
```

```python
data['subject'] = pd.Categorical(
    data['subject'],
    ['cpython', 'pypy', 'numba', 'pyston-lite', 'py3.13-jit', 'nuitka', 'cython', 'codon', 'mypyc']
)
```

```python
def improvement(old_values, new_values):
    improvements = []
    for old, new in zip(old_values, new_values):
        if old == 0:
            improvements.append(None)
        else:
            improvement = ((old - new) / old) * 100
            improvements.append(improvement)
    return improvements
```

```python
def speedup(old_values, new_values):
    speedups = []
    for old, new in zip(old_values, new_values):
        if old == 0:
            speedups.append(None)
        else:
            speedup = old / new
            speedups.append(speedup)
    return speedups
```

```python
#subjects = list(set(data['subject'].values))
subjects = ['cpython', 'pypy', 'numba', 'pyston-lite', 'py3.13-jit', 'nuitka', 'cython', 'codon', 'mypyc']
targets = ['mandelbrot', 'fannkuch_redux', 'binary_trees', 'n_body', 'spectralnorm', 'k_nucleotide', 'fasta']

GROUPS = list(
    product(list(set(subjects) - {'cpython'}), targets)
)
```

```python
def calc_stat(data, fun):
    from statistics import mean
    rows = []
    for x, y in GROUPS:
        baseline = data[(data['subject'] == 'cpython') & (data['target'] == y)][METRIC].to_list()
        new_values = data[(data['subject'] == x) & (data['target'] == y)][METRIC].to_list()
    
        row = {'subject': x , 'target' : y, METRIC : statistics.fmean(fun([mean(baseline)], [mean(new_values)]))}
        rows.append(row)
        
    return pd.DataFrame(rows)
```

```python
def plot_stat(data):
    colors = [
        "#FF6B6B",  # Coral Red
        "#4ECDC4",  # Turquoise
        "#FFA07A",  # Light Salmon
        "#6A3D9A",  # Purple
        "#FFD700",  # Gold
        "#45B7D1",  # Sky Blue
        "#98D8C8",  # Mint
        "#FF69B4",  # Hot Pink
        "#8B4513"   # Saddle Brown
    ]
    #Convert 'target' column to categorical data type with custom order
    data['target'] = pd.Categorical(data['target'], categories=targets, ordered=True)
    
    # Sort the dataframe based on the custom order
    data = data.sort_values('target')
    
    # Create the lineplot
    sns.set_theme(style="whitegrid", context="talk")
    
    fig, ax = plt.subplots(figsize=(10, 8), dpi=300)
    
    plt.rcParams.update({
        'font.size': 14,
        'axes.labelsize': 16,
        'axes.titlesize': 18,
        'xtick.labelsize': 18,
        'ytick.labelsize': 18,
        'legend.fontsize': 12
    })
    
    plt.tick_params(axis='both', which='major', labelsize=18)
    
    sns.lineplot(
        data=data, 
        x='target', y=METRIC, hue='subject', linewidth=3, marker="o", markersize=8, palette=colors
    )
    #sns.despine()
    
    plt.title('')
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=45)
    plt.yticks()
    
    plt.legend().remove()
    plt.tight_layout()
    
    plt.savefig(f"{SERVER}_{METRIC}_improvement.pdf", dpi=300)
    plt.show()
```

```python
improvement_data = calc_stat(data, improvement)
```

```python
improvement_data['subject'] = pd.Categorical(
    improvement_data['subject'],
    ['cpython', 'pypy', 'numba', 'pyston-lite', 'py3.13-jit', 'nuitka', 'cython', 'codon', 'mypyc']
)

improvement_data = improvement_data.sort_values(by='subject')
```

```python
plot_stat(improvement_data)
```

```python
METRIC = 'execution_time'
speedup_data = calc_stat(data, speedup)
speedup_data['subject'] = pd.Categorical(
    speedup_data['subject'],
    ['cpython', 'pypy', 'numba', 'pyston-lite', 'py3.13-jit', 'nuitka', 'cython', 'codon', 'mypyc']
)

speedup_data = speedup_data.sort_values(by='subject')
```

```python
plot_stat(speedup_data)
```
